```python
import pandas as pd
from numpy.random import choice
import random
```

```python
cluster=pd.read_csv("Cluster Weighting.csv")
matrix=pd.read_csv("Matrix Score.csv")
# user inputs will pull from database
user_inputs = pd.read_csv("user inputs from database.csv")

# get cluster from one or two questions
user_cluser = "Cluster 2"

print(matrix.head())

print(cluster.head())

print(user_inputs.head())
```

```python
# merge matrix with user answer 
user_scale = matrix.merge(user_inputs, left_on = 'Question', right_on='Questions')
user_scale.head()
```

```python
# clean up the merge dataframe and filter out the values only for answers
user_scale = user_scale.loc[user_scale['Question_Scale'] == user_scale['Answer']]
user_scale
```

```python
user_scale_total = user_scale.groupby(['Category']).sum(['Value']).drop(columns = ['Question_Scale','Answer'])
user_scale_total
```

```python
# Merge Cluster with user_scale_total
user_final = cluster.loc[cluster['Cluser']== user_cluser]
user_final_Cat = user_final.merge(user_scale_total, how = "left", left_on = "Category", right_on= "Category")
user_final_Cat['Final_Score'] = user_final_Cat['Value'] * user_final_Cat['Weight']
user_final_Cat
```

```python
activity = user_final_Cat.loc[user_final_Cat['Group'] == 'activity']
activity_per = activity['Final_Score']/activity['Final_Score'].sum()
act_choice = choice(activity['Category'],p=activity_per)

food = user_final_Cat.loc[user_final_Cat['Group'] == 'food&drink']
food_per = food['Final_Score']/food['Final_Score'].sum()
food_choice = choice(food['Category'],p=food_per)

sight = user_final_Cat.loc[user_final_Cat['Group'] == 'sightseeing']
sight_per = sight['Final_Score']/sight['Final_Score'].sum()
sight_choice = choice(sight['Category'],p=sight_per)



# user_final_Cat['Total_Score__Cat'] = user_final_Cat.groupby('Group').sum(['Final_Score'])
# top_3 = user_final_Cat.loc[user_final_Cat.groupby('Group')["Final_Score"].idxmax()]
# top_3 = list(top_3["Category"])
# top_3
top_3 = [act_choice,food_choice,sight_choice]
top_3
```
