```python
import pandas as pd
import csv
import os
import matplotlib as plt
```

```python
# other clustering, path:
# PCA_BK\clustered_google_data_1.csv
# PCA_BK_NoLS\clustered_gd_noLS.csv
# PCA_BK_NoLocalSHotel\clustered_gd_noLSH.csv
# path could be improved

cluster_ratings = pd.read_csv("..\PCA_BK_NoLocalSHotel\clustered_gd_noLSH.csv")
```

```python
cluster_ratings.head()
```

```python
#Clean the data

# Rename columns to appropriate place `type` 

cluster_ratings = cluster_ratings.rename({'churches': 'place_of_worship',
                                                 'resorts': 'resort',
                                                 'beaches': 'beach',
                                                 'parks': 'park',
                                                 'theatres': 'movie_theater',
                                                 'museums': 'museum',
                                                 'malls': 'shopping_mall',
                                                 'restaurants': 'restaurant',
                                                 'pubs-bars': 'bar',
                                                 'art-galleries': 'art_gallery',
                                                 'dance-clubs': 'night_club',
                                                 'gyms': 'gym',
                                                 'beauty-spas': 'beauty-hair-spa',
                                                 'bakeries': 'bakery',
                                                 'cafes': 'cafe',
                                                 'monuments': 'monument'
                                                 }, axis=1)

#Drop 5-cluster
cluster_ratings = cluster_ratings.drop('user_id', axis=1)
cluster_ratings = cluster_ratings.drop('Clusters5', axis=1)

cluster_ratings.head(5)
```

```python
# find the mean of ratings by cluster and combine
cluster_0 = cluster_ratings[cluster_ratings.Clusters4 == 0].mean(axis=0,skipna=True)
cluster_1 = cluster_ratings[cluster_ratings.Clusters4 == 1].mean(axis=0,skipna=True)
cluster_2 = cluster_ratings[cluster_ratings.Clusters4 == 2].mean(axis=0,skipna=True)
cluster_3 = cluster_ratings[cluster_ratings.Clusters4 == 3].mean(axis=0,skipna=True)
```

```python
weight = pd.DataFrame([cluster_0, cluster_1, cluster_2, cluster_3])
weight = weight.rename(index={0: "cluster0", 1: "cluster1", 2: "cluster2", 3:"cluster3"})
weight = weight.drop(labels='Clusters4', axis=1)
weight.head()
```

```python
weight.style.background_gradient(cmap='YlGn', high=2)
```

```python
# write the weights to .csv
os.makedirs('Output', exist_ok=True)
weight.to_csv('Output/weights.csv')
```
