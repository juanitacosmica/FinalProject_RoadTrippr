# Clustering the google_review_ratings dataset

Testing the ability for clustering algorithms to categorize chosen

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
grr = pd.read_csv('google_review_ratings2.csv', index_col=False) 
```

```python
grr.head()
```

```python
# Rounding ALL categories to integers for simplicity of processing and comprehending.
import numpy as np
grr2 = np.round(grr, decimals=0)
grr2.index += 1 
grr2.tail()
```

is_zero =  grr2==0.0
is_one = grr2==1.0
is_two = grr2==2.0
is_three = grr2==3.0
is_four = grr2==4.0
is_five = grr2==5.0

zero_frame = grr2[is_zero]
one_frame = grr2[is_one]
two_frame = grr2[is_two]
three_frame = grr2[is_three]
four_frame = grr2[is_four]
five_frame = grr2[is_five]


five_frame.head()

```python
five_frame = grr2.apply(lambda row: row[row == 5.0].index.tolist(), axis=1)
four_frame = grr2.apply(lambda row: row[row == 4.0].index.tolist(), axis=1)
three_frame = grr2.apply(lambda row: row[row == 3.0].index.tolist(), axis=1)
two_frame = grr2.apply(lambda row: row[row == 2.0].index.tolist(), axis=1)
one_frame = grr2.apply(lambda row: row[row == 1.0].index.tolist(), axis=1)
zero_frame = grr2.apply(lambda row: row[row == 0.0].index.tolist(), axis=1)
```

```python
grr3 = pd.DataFrame(five_frame.values)
grr3 = grr3.merge(pd.DataFrame(four_frame.values),left_index=True, right_index=True)
grr3 = grr3.merge(pd.DataFrame(three_frame.values),left_index=True, right_index=True)
grr3 = grr3.merge(pd.DataFrame(two_frame.values),left_index=True, right_index=True)
grr3 = grr3.merge(pd.DataFrame(one_frame.values),left_index=True, right_index=True)
grr3 = grr3.merge(pd.DataFrame(zero_frame.values),left_index=True, right_index=True)
grr3.index += 1 
grr3.columns = ['five_stars', 'four_stars', 'three_stars', 'two_stars', 'one_star', 'zero_star']
```
